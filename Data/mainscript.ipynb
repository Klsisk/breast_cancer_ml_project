```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

# Extract CSVs into DataFrames

```python
# Read Breast Cancer csv and store into Pandas data frame
df ="breast_cancer.csv"
df = pd.read_csv(df)
df.head()
```

# Transform Salary DataFrame

```python
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
# Drop the column id
df=df.iloc[:,1:]
df.head()
```

```python
df[df['bare_nucleoli']=='?'].sum()
```

```python
digits_in_bare_nucleoli=pd.DataFrame(df.bare_nucleoli.str.isdigit())

digits_in_bare_nucleoli
```

```python
data=df.replace("?", np.nan)
data.head()
```

# Create database connection

```python
engine = create_engine(f'postgresql://postgres:{sql_pass}@localhost/breast_cancer')
```

```python
#Confirm tables
engine.table_names()
```

# Load DataFrames into database

```python
data.to_sql(name='cancerData', con=engine, if_exists='append', index=False)
```
