```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

# Extract CSVs into DataFrames

# Read Breast Cancer csv and store into Pandas data frame

```python
df ="breast_cancer.csv"
df = pd.read_csv(df)
df.head()
```

```python
# Transform Salary DataFrame
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all').copy()

# Drop the null rows
df = df.dropna()

# Drop the column id
df.head()
```

```python
df.columns[0:12]
```

```python
df = df[df.columns[1:12]]
df
```

```python
#Get Datatypes of each column in our dataset
df.dtypes
```

```python
#Encode the diagnosis values
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df.iloc[:,0]=labelencoder.fit_transform(df.iloc[:,0].values)
```

```python
# Check datatypes again
df.dtypes
```

# Create database connection

```python
engine = create_engine(f'postgresql://postgres:{sql_pass}@localhost/breast_cancer1')
```

```python
#Confirm tables
engine.table_names()
```

# Load DataFrames into database

```python
df.to_sql(name='cancerdata', con=engine, if_exists='append', index=False)
```
